# concatenating 

```python
import pandas as pd

def concat_dataframes(n=48):
    """
    Concatenate DataFrames named df1, df2, ..., dfN into a single DataFrame.

    Parameters:
        n (int): The number of DataFrames to check (default = 48).
                 For example, if n=48, it looks for df1, df2, ..., df48.

    Returns:
        pd.DataFrame: A concatenated DataFrame containing all the found DataFrames.
                      Returns an empty DataFrame if none are found.
    """
    
    dfs = []  # this list will store all the found DataFrames
    
    # Loop from 1 to n (inclusive)
    for i in range(1, n+1):
        var_name = f"df{i}"  # construct variable name like 'df1', 'df2', etc.
        
        # Check if this variable exists in the current global namespace
        if var_name in globals():
            dfs.append(globals()[var_name])  # add the DataFrame to our list
        else:
            print(f" Skipping {var_name} (not found)")  # warning if DataFrame doesn't exist
    
    # After collecting all available DataFrames
    if dfs:  # only concatenate if the list is not empty
        main_df = pd.concat(dfs, ignore_index=True)  # stack them together into one DataFrame
        return main_df
    else:
        print(" No DataFrames found to concatenate.")
        return pd.DataFrame()  # return empty DataFrame if nothing found


#  Usage
main_df = concat_dataframes(48)  # will try to concatenate df1 ... df48
```
