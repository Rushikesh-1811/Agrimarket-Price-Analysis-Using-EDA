# Data Cleaning

## Replacing all blanks with '_' in all column names

```python
df.columns =(
    df.columns
    .str.strip()
    .str.replace(r'\s+', '_', regex=True)
    .str.replace(r'[^A-Za-z0-9_]', '', regex=True)
)
df.rename(columns={
    df.columns[0]: 'State_Name',
    df.columns[1]: 'District_Name',
    df.columns[2]: 'Market_Name',
    df.columns[3]: 'Variety',
    df.columns[4]: 'Group',
    df.columns[5]: 'Arrivals_(Tonnes)',
    df.columns[6]: 'Min_Price_(Rs./Quintal)',
    df.columns[7]: 'Max_Price_(Rs./Quintal)',
    df.columns[8]: 'Modal_Price_(Rs./Quintal)',
    df.columns[9]: 'Reported_Date',
    df.columns[10]: 'Grade'
}, inplace=True)

df.columns
```

```python
df.head()
```

# Replace multiple patterns in column names programmatically using RegEx

```python
df.columns = df.columns.str.replace(r"\(|\)|\.","",regex  = True).str.replace("/","_")
```

```python
df['Variety'] = df['Variety'].replace('Red', 'Red Onion')
```

# Dropping unwanted rows(SubTotal / Total)

```python
df = df[~df["District_Name"].str.contains("Total",na = False)]
```

# converting Report Date column to datetime

```python
df['Reported Date'] = pd.to_datetime(
    df['Report_Date'].str.strip(),  # remove extra spaces
    format="%d %b %Y",  # example: "03 Jun 2025"
    errors="coerce"     # still gives NaT if mismatch
)
```

```python
df.drop('Report_Date', axis=1, inplace=True)
```
